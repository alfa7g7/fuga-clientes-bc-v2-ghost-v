```python
# ============== Interpretacion_modelo_fuga_clientes.ipynb ==============

# --- 1. Importaciones Esenciales ---
import pandas as pd
import numpy as np
import joblib
import pickle
import os
import re # Para Rename_columns
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor # Para Ghost y RF de referencia
from sklearn.preprocessing import LabelEncoder, RobustScaler, OneHotEncoder, StandardScaler # Para Ghost y RF de referencia
from sklearn.impute import SimpleImputer, KNNImputer # Para Ghost y RF de referencia
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from sklearn.linear_model import LassoCV # Para el SelectFromModel
from sklearn.feature_selection import SelectFromModel
from xgboost import XGBClassifier # Para la definición del pipeline si es necesario

# >>> AÑADIR ESTO SI TRABAJAS EN COLAB Y LOS ARCHIVOS ESTÁN EN DRIVE <<<
from google.colab import drive
drive.mount('/content/drive')

# --- 2. Definición de TODAS las Clases Transformer Personalizadas ---
# Check_And_Remove_New_Lines_Spaces, Rename_columns, DateTime_Feature_Extractor,
# Convert_Variable_To_Float, Convert_Variable_To_Categorical,
# Convert_Mixed_Types_To_Str, Format_Categorical_Variables,
# Remove_custom_features, Drop_columns_by_missing_Values, Age_Filter (versión INTERNA pipeline)

class Check_And_Remove_New_Lines_Spaces(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        """
        Verifica si los registros en un DataFrame contienen saltos de línea y permite eliminarlos.
        También elimina los espacios al inicio y al final de cada registro.

        Parámetros:
        - X: DataFrame que contiene los datos.

        Retorna:
        - Un DataFrame con los saltos de línea eliminados de los registros y sin espacios al inicio o al final.
        """
        X = X.copy()

        # Verificar si hay saltos de línea en cada celda del DataFrame
        newline_positions = {}
        for col in X.columns:
            newline_positions[col] = X[col].apply(lambda x: '\n' in str(x))

        # Mostrar las posiciones de los saltos de línea
        for col, positions in newline_positions.items():
            if positions.any():
                print(f"Saltos de línea encontrados en la columna '{col}':")
                print(X[positions][col])
                print("\n")

        # Eliminar los saltos de línea y los espacios al inicio y al final
        for col in X.columns:
            X[col] = X[col].apply(lambda x: str(x).replace('\n', ' ').strip() if isinstance(x, str) else x)

        return X

class Rename_columns(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        result = X.copy()

        # Aplicamos el formato: eliminamos espacios, convertimos a minúsculas y cambiamos espacios medios por guion bajo
        new_cols = []
        for col in result.columns:
            new_cols.append(re.sub(r'\s+', ' ', col.strip()).replace(' ', '_').lower())

        # Asignamos las nuevas columnas al DataFrame
        result.columns = new_cols

        # Renombramos columnas específicas si es necesario
        result = result.rename(columns={'#': 'id_anon',
                                        'clase_fuga': 'fuga_vo',
                                        'fecha_fin': 'fecha_ultimo_vista',
                                        'fecha_valor': 'fecha_desembolso',
                                        'fechadevencimiento': 'fecha_de_vencimiento',
                                        'importeoriginal': 'monto_desembolsado',
                                        'saldoactual': 'saldo_ultima_vista',
                                        'díasdemora': 'dias_de_mora',
                                        'diasantiguedad': 'dias_antiguedad',
                                        'permanencia': 'permanencia_meses',
                                        'nombre_suc': 'nombre_sucursal',
                                        'region,1': 'region',
                                        'zona,1': 'zona',
                                        'departamento,negocio,empresa': 'dpto_negocio',
                                        'plazooperación': 'plazo_operacion_dias',
                                        'periodicidad': 'periodicidad_pago_dias',
                                        'tasaoriginal': 'tasa_al_desembolso',
                                        'tasaactual': 'tasa_ultima_vista',
                                        'anti': 'tipo_cliente',
                                        'tipo,crédito': 'tipo_credito',
                                        'tipo,vivienda': 'tipo_vivienda',
                                        'estado,civil': 'estado_civil',
                                        'cod_ciiu': 'codigo_ciiu',
                                        'desc_ciiu': 'descripcion_ciiu',
                                        'desc_niveleducacion': 'nivel_educacion',
                                        'desc_ocupacion': 'ocupacion',
                                        'total_ingresosmensuales': 'total_ingresos_mensuales',
                                        'total_egresosmensuales': 'total_egresos_mensuales',
                                        })
        return result


class DateTime_Feature_Extractor(BaseEstimator, TransformerMixin):
    def __init__(self, columns, convert_columns=None, check_columns=None, drop_original=True):
        """
        Transformador para convertir columnas a datetime y extraer características (año, mes, día) de columnas datetime existentes.
        Adicionalmente, elimina registros con valores nulos o cero en las columnas especificadas.

        Parámetros:
        - columns: lista de columnas de las que extraer características.
        - convert_columns: lista de columnas a convertir a datetime.
        - check_columns: lista de columnas a verificar para valores nulos o cero.
        - drop_original: bool, si True elimina las columnas originales después de extraer las características.
        """
        self.columns = columns
        self.convert_columns = convert_columns
        self.check_columns = check_columns
        self.drop_original = drop_original

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        X = X.copy()

        # Convertir columnas a datetime
        if self.convert_columns:
            for col in self.convert_columns:
                if col not in X.columns:
                    raise ValueError(f"La columna '{col}' no se encontró en los datos de entrada.")
                X[col] = pd.to_datetime(X[col], errors='coerce')

        # Extraer características de fecha
        for col in self.columns:
            if col not in X.columns:
                raise ValueError(f"La columna '{col}' no se encontró en los datos de entrada.")

            if not pd.api.types.is_datetime64_any_dtype(X[col]):
                raise TypeError(f"La columna '{col}' no es de tipo datetime.")

            X[f'{col}_año'] = X[col].dt.year.astype('float64')
            X[f'{col}_mes'] = X[col].dt.month.astype('float64')
            X[f'{col}_dia'] = X[col].dt.day.astype('float64')

            if self.drop_original:
                X = X.drop(columns=[col])

        # Imprimir registros con valores nulos o cero en las columnas especificadas
        if self.check_columns:
            for columna in self.check_columns:
                if columna in X.columns:
                    nulos_ceros = X[(X[columna].isnull()) | (X[columna] == 0)]
                    if not nulos_ceros.empty:
                        print(f"Registros con valores nulos o cero en la columna '{columna}':")
                        print(nulos_ceros)
                        print("\n")


            # Eliminar registros con valores nulos o cero en las columnas especificadas
            X = X.dropna(subset=self.check_columns)
            for columna in self.check_columns:
                if columna in X.columns:
                    X = X[X[columna] != 0]


        return X

class Convert_Variable_To_Float(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        for col in self.columns:
            X[col] = X[col].replace(r'^\s*$', 0, regex=True)  # Replace empty strings with zeros
            X[col] = pd.to_numeric(X[col], errors='coerce').fillna(0)  # Convert to numeric and replace non-numeric with zeros
            X[col] = X[col].astype(float)
        return X

class Convert_Variable_To_Categorical(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        for col in self.columns:
            X[col] = X[col].astype('category')
        return X

class Convert_Mixed_Types_To_Str(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        for col in X.columns:
            if X[col].apply(lambda x: isinstance(x, str)).any() and X[col].apply(lambda x: isinstance(x, (int, float))).any():
                X[col] = X[col].astype(str)
        return X

class Format_Categorical_Variables(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        result = X.copy()

        # Iterar sobre todas las columnas categóricas del DataFrame
        for col in result.select_dtypes(include=['object', 'category']).columns:
            #print(f"Procesando columna: {col}")
            # Convertir todos los valores a cadenas de texto
            #result[col] = result[col].astype(str)

            # Eliminar espacios al principio y al final, y eliminar saltos de línea
            #result[col] = result[col].str.strip().str.replace('\n', '', regex=False).str.replace('\r\n', '', regex=False)

            # Diseñamos algunas reglas específicas:
            result[col] = result[col].replace({
                'NORMAL': 'Normal',  # Variable Tipo_credito
                'FAMILIAR': 'Familiar',  # Variable tipo_vivienda
                'nan': np.nan,  # Convertir 'nan' a NaN
                '0': np.nan,  # Convertir '0' a NaN
                'UNIÓN LIBRE': 'Unión Libre',  # Variable Estado_civil
                'Separado/a': 'Separado/Divorciado',  # Variable Estado_civil
                'Divorciado/a': 'Separado/Divorciado',  # Variable Estado_civil
                '': np.nan  # Convertir cadenas vacías a NaN
            })

            # Imprimir los primeros valores de la columna después de la transformación
            #print(result[col].head())

        return result

class Remove_custom_features(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        result = X.copy()

        # Verificar si las columnas existen antes de eliminarlas
        columns_to_drop = ['region', 'zona']
        existing_columns_to_drop = [col for col in columns_to_drop if col in result.columns]

        if existing_columns_to_drop:
            result = result.drop(columns=existing_columns_to_drop)

        return result

class Drop_columns_by_missing_Values(BaseEstimator, TransformerMixin):
    def __init__(self, threshold=0.75):  # Umbral en porcentaje, 0.75 = 75%
        self.threshold = threshold
        self.columns_to_drop = []

    def fit(self, X, y=None):
        # Calcular el porcentaje de valores faltantes por columna
        percent_missing = X.isnull().mean()
        # Guardar las columnas que superan el umbral
        self.columns_to_drop = percent_missing[percent_missing > self.threshold].index
        return self

    def transform(self, X, y=None):
        # Eliminar las columnas que superan el umbral
        return X.drop(columns=self.columns_to_drop)

class Age_Filter(BaseEstimator, TransformerMixin): # Asegúrate que es la versión interna; para PIPELINE
    def __init__(self, reference_year=2024, min_age=18, max_age=80):
        self.reference_year = reference_year
        self.min_age = min_age
        self.max_age = max_age
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        if isinstance(X, tuple): X = X[0]
        result = X.copy()
        result['age_calc'] = self.reference_year - result['fecha_nacimiento_año']
        mask = (result['age_calc'] >= self.min_age) & (result['age_calc'] <= self.max_age)
        result = result[mask]
        if y is not None:
            if isinstance(y, pd.Series): y = y[mask]
            elif isinstance(y, np.ndarray): y = y[mask]
        result = result.drop(columns=['age_calc'])
        if y is not None: return result, y
        return result

# --- 3. Cargar Modelo Pre-entrenado ---
model_path = '/content/drive/MyDrive/CDD/Sem02/PGI/proyecto-grado-fuga-clientes-bc/models/best_model_xgb_v2.pkl'
if os.path.exists(model_path):
    pipeline_cargado = joblib.load(model_path)
    print(f"Modelo cargado exitosamente desde: {model_path}")
else:
    print(f"Error: No se encontró el archivo del modelo en {model_path}")
    pipeline_cargado = None
    exit() # Salir si el modelo no se puede cargar

# --- 4. Cargar y Preparar los Datos de Evaluación ---
# (X_test, y_test) - Los mismos que se usaron para la evaluación final del modelo guardado
# Es crucial que estos datos estén en su forma ANTES de pasar por el pipeline_cargado
# Es decir, como estaban después del train_test_split inicial.

ruta_archivo_original = '/content/drive/MyDrive/CDD/Sem03/PGII/fuga-clientes-bc-v2-ghost-v/Data/Base_Fuga_24.xlsx'
df_original = pd.read_excel(ruta_archivo_original)

# Aplicar transformaciones PRE-PIPELINE que hiciste antes del split en el notebook original
df_temp = Rename_columns().transform(df_original)
df_temp = DateTime_Feature_Extractor(
    columns=['fecha_desembolso', 'fecha_ultimo_vista', 'fecha_de_vencimiento', 'fecha_nacimiento'],
    convert_columns=['fecha_nacimiento'],
    check_columns=['fecha_desembolso_año', 'fecha_nacimiento_año']
).fit_transform(df_temp)

# Para Age_Filter externa, necesitas el df_temp antes de instanciar
age_filter_externo = Age_Filter(reference_year=int(df_temp['fecha_desembolso_año'].max()))
df_temp = age_filter_externo.fit_transform(df_temp)


# Recrear los splits EXACTAMENTE como en el notebook original
df_entrenamiento_completo, df_val_final = train_test_split(df_temp, test_size=0.08, random_state=42, stratify=df_temp['fuga_vo'])
X_original = df_entrenamiento_completo.drop('fuga_vo', axis=1)
y_original = df_entrenamiento_completo['fuga_vo']
X_train_orig, X_test_eval, y_train_orig, y_test_eval = train_test_split(X_original, y_original, test_size=0.3, random_state=42, stratify=y_original)

print(f"Tamaño de X_test_eval para análisis fantasma: {X_test_eval.shape}")

# Definir las listas de variables (tal como las tenías)
var_numericas = ['total_ingresos_mensuales', 'total_egresos_mensuales', 'codigo_ciiu', 'total_activos', 'dias_antiguedad', 'total_pasivo', 'plazo_operacion_dias', 'saldo_ultima_vista', 'monto_desembolsado', 'id_anon', 'permanencia_meses', 'fecha_nacimiento_dia', 'fecha_nacimiento_año', 'fecha_desembolso_dia', 'calificacion', 'fecha_desembolso_año', 'fecha_de_vencimiento_dia', 'cantidad_hijos', 'fecha_desembolso_mes', 'fecha_ultimo_vista_mes', 'fecha_nacimiento_mes', 'tasa_al_desembolso', 'tasa_ultima_vista', 'fecha_de_vencimiento_mes']
var_categoricas = ['descripcion_ciiu', 'centro_poblado', 'tipo_cliente', 'nombre_sucursal', 'sector', 'dpto_negocio', 'zona', 'tipo_credito', 'tipo_vivienda', 'nivel_educacion', 'region']

# Asegurarse que `pipeline_cargado` no es None
if pipeline_cargado is None:
    print("El pipeline no se cargó, deteniendo ejecución.")
    # exit() o raise Exception

# --- 5. Definición de la Función `relev_ghost_var` ---
def relev_ghost_var(model_pipeline, X_data, y_data, features_list, func_model_ghost_var=RandomForestRegressor, n_estimators_ghost=100, random_state_ghost=123):
    """
    Calcula la relevancia de las variables usando el método de Variables Fantasma.

    Args:
        model_pipeline: El pipeline de scikit-learn entrenado (incluye preprocesamiento y modelo final).
        X_data (pd.DataFrame): DataFrame con las características (datos de test o validación).
        y_data (pd.Series): Serie con la variable objetivo (datos de test o validación).
        features_list (list): Lista de nombres de las características a evaluar.
        func_model_ghost_var: Clase del modelo a usar para generar las variables fantasma
                               (e.g., RandomForestRegressor, RandomForestClassifier).
        n_estimators_ghost (int): Número de estimadores para el modelo fantasma (si es RF).
        random_state_ghost (int): Semilla aleatoria para el modelo fantasma.

    Returns:
        dict: Diccionario con la relevancia calculada ('relev_ghost'), la matriz V,
              los vectores de efecto A, las predicciones base y las predicciones ghost.
    """
    features_list = var_numericas + var_categoricas # Lista de nombres de las características
    term_labels = features_list # Usamos la lista de features proporcionada
    p = len(term_labels)
    n2 = X_data.shape[0]

    # --- Pre-procesamiento Básico para Modelos Fantasma ---
    # Crear una copia para no modificar el original
    X_ghost_prep = X_data[term_labels].copy()

    # Codificar variables categóricas temporalmente para modelos fantasma (si usan RF)
    # Nota: Una imputación simple podría ser necesaria aquí si RF no maneja NaNs
    encoders = {}
    for col in X_ghost_prep.select_dtypes(include=['object', 'category']).columns:
        # Imputar NaNs categóricos con la moda antes de codificar
        if X_ghost_prep[col].isnull().any():
             mode_val = X_ghost_prep[col].mode()[0]
             X_ghost_prep[col] = X_ghost_prep[col].fillna(mode_val)

        # Usar LabelEncoder simple para RF (RF puede manejar enteros como categorías)
        le = LabelEncoder()
        X_ghost_prep[col] = le.fit_transform(X_ghost_prep[col].astype(str)) # Asegurar que sea string
        encoders[col] = le # Guardar encoder si se necesita revertir (no necesario aquí)

    # Imputar NaNs numéricos con la media para modelos fantasma
    for col in X_ghost_prep.select_dtypes(include=np.number).columns:
         if X_ghost_prep[col].isnull().any():
             mean_val = X_ghost_prep[col].mean()
             X_ghost_prep[col] = X_ghost_prep[col].fillna(mean_val)

    # --- Calcular Predicciones Base ---
    # Usar el pipeline completo para predecir probabilidades en los datos originales
    try:
        y_hat_ts = model_pipeline.predict_proba(X_data)[:, 1] # Probabilidad clase 1 (Fuga)
    except Exception as e:
        print(f"Error prediciendo con el pipeline principal: {e}")
        print("Asegúrate que el pipeline esté entrenado y que X_data tenga el formato correcto.")
        return None

    A = np.zeros((n2, p)) # Matriz de efectos (diferencia en predicciones)
    GhostX_pred = np.zeros((n2, p)) # Almacenar las predicciones fantasma (xj_hat)

    # --- Calcular Relevancia para cada variable ---
    for j, var in enumerate(term_labels):
        print(f"Calculando ghost para variable: {var} ({j+1}/{p})")
        X_ghost_prep_aux = X_ghost_prep.copy() # Copia para esta iteración
        X_data_aux = X_data.copy() # Copia de los datos originales para predicción

        # Definir predictores para el modelo fantasma (todas las otras variables)
        predictors = [label for label in term_labels if label != var]
        if not predictors: # Si solo hay una variable
             print(f"Skipping ghost model for {var}, no other predictors.")
             xj_hat = X_ghost_prep_aux[var].mean() # Usar la media como predicción simple
             xj_hat = np.full(n2, xj_hat) # Crear array
        else:
            # Seleccionar el tipo de modelo fantasma basado en el tipo de variable 'var'
            is_categorical = var in X_data.select_dtypes(include=['object', 'category']).columns

            if is_categorical:
                # Variable categórica: Usar clasificador
                ghost_model = RandomForestClassifier(n_estimators=n_estimators_ghost, random_state=random_state_ghost, n_jobs=-1)
                target_ghost = X_ghost_prep_aux[var] # Ya está codificada numéricamente
            else:
                 # Variable numérica: Usar regresor
                 ghost_model = RandomForestRegressor(n_estimators=n_estimators_ghost, random_state=random_state_ghost, n_jobs=-1)
                 target_ghost = X_ghost_prep_aux[var]

            # Entrenar el modelo fantasma
            # Asegurarse que no haya NaNs en los predictores para el modelo fantasma
            X_pred_ghost = X_ghost_prep_aux[predictors]
            if X_pred_ghost.isnull().values.any():
                 print(f"Warning: NaNs found in predictors for ghost model of {var}. Check imputation.")
                 # Aplicar imputación simple si aún hay NaNs (debería haberse hecho antes)
                 for pred_col in predictors:
                     if X_pred_ghost[pred_col].isnull().any():
                         if X_pred_ghost[pred_col].dtype in ['float64', 'int64']:
                              X_pred_ghost[pred_col] = X_pred_ghost[pred_col].fillna(X_pred_ghost[pred_col].mean())
                         else: # Categorical already encoded
                              X_pred_ghost[pred_col] = X_pred_ghost[pred_col].fillna(X_pred_ghost[pred_col].mode()[0])

            ghost_model.fit(X_pred_ghost, target_ghost)
            xj_hat = ghost_model.predict(X_pred_ghost) # Predicción fantasma

        GhostX_pred[:, j] = xj_hat # Guardar predicción fantasma

        # Reemplazar la variable original con su fantasma en la copia de datos ORIGINALES
        # Si la variable original era categórica, necesitamos mapear la predicción fantasma (que es numérica/código) de vuelta,
        # o si el OHE del pipeline puede manejarla directamente. Lo más seguro es que falle si espera strings.
        # Aquí asumimos que el reemplazo directo funciona si el pipeline maneja tipos mixtos o si la variable es numérica.
        # Si 'var' es categórica, xj_hat son códigos. El pipeline necesita la categoría original. ¡Esto es complejo!
        # SOLUCIÓN MÁS SIMPLE (Compromiso): Si var es categórica, el ghost model predice la *moda* o una categoría fija.
        # OMITAMOS categóricas por ahora para simplificar, como en el ejemplo original.

        if var in X_data.select_dtypes(include=['object', 'category']).columns:
             print(f"Skipping prediction for categorical ghost variable '{var}' due to complexity.")
             # Opcional: asignar un efecto nulo o basado en predicción con moda
             A[:, j] = 0 # Asignar efecto cero temporalmente
             continue # Saltar al siguiente bucle

        # Reemplazar en la copia de datos originales (para el pipeline principal)
        X_data_aux[var] = xj_hat

        # Predecir usando el pipeline principal con la variable fantasma
        try:
             y_hat_ts_j = model_pipeline.predict_proba(X_data_aux)[:, 1]
        except Exception as e:
             print(f"Error prediciendo con ghost variable {var}: {e}")
             print("Puede haber un problema de tipo de dato al reemplazar con xj_hat.")
             A[:, j] = 0 # Asignar efecto cero si falla
             continue

        # Calcular el vector de efecto
        A[:, j] = y_hat_ts - y_hat_ts_j

    # Calcular la matriz V (sin normalizar por MSPE)
    # Manejar el caso donde n2 es 0
    if n2 > 0:
        V = (1 / n2) * np.dot(A.T, A)
    else:
        V = np.zeros((p, p)) # O manejar como error

    # Relevancia individual como la diagonal de V
    relev_ghost = np.diag(V)

    # Asegurarse que no haya valores negativos (pueden ocurrir por errores numéricos)
    relev_ghost = np.maximum(relev_ghost, 0)

    return {
        "A": A,
        "V": V,
        "GhostX_pred": GhostX_pred, # Las predicciones xj_hat
        "relev_ghost": relev_ghost, # Relevancia individual (diagonal de V)
        #"eig_V": np.linalg.eig(V), # Descomposición propia (opcional)
        "y_hat_base": y_hat_ts # Predicciones base
    }

# --- 6. Ejecutar Análisis de Variables Fantasma ---
features_to_evaluate = var_numericas + var_categoricas # Lista COMPLETA para que la función filtre

sample_size = 5000
if len(X_test_eval) > sample_size:
    sample_indices = np.random.choice(X_test_eval.index, size=sample_size, replace=False)
    X_test_sample = X_test_eval.loc[sample_indices]
    y_test_sample = y_test_eval.loc[sample_indices]
else:
    X_test_sample = X_test_eval
    y_test_sample = y_test_eval

print(f"Usando muestra de tamaño: {len(X_test_sample)} para análisis fantasma.")

relev_ghost_output = relev_ghost_var(pipeline_cargado, X_test_sample, y_test_sample, features_to_evaluate)

# --- 7. Definir `num_importances` (Cargarla o Recalcularla como fallback) ---
# Intenta cargarla primero
num_importances_path = '/content/drive/MyDrive/CDD/Sem03/PGII/fuga-clientes-bc-v2-ghost-v/num_importances/num_importances_rf_eda.pkl' #Ajusta la ruta
if os.path.exists(num_importances_path):
    with open(num_importances_path, 'rb') as f:
        num_importances = pickle.load(f)
    print("num_importances cargada desde archivo.")
else:
    print("Archivo num_importances_rf_eda.pkl no encontrado. Recalculando...")
    # Recalcular num_importances (como en el bloque anterior)
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.impute import SimpleImputer
    num_features_for_rf = [f for f in var_numericas if f in X_test_sample.columns] # Usar var_numericas
    if num_features_for_rf:
        X_num_sample_rf = X_test_sample[num_features_for_rf].copy()
        imputer_rf = SimpleImputer(strategy='mean')
        X_num_sample_rf_imputed = imputer_rf.fit_transform(X_num_sample_rf)
        rf_model_imp = RandomForestClassifier(n_estimators=1000, random_state=42, n_jobs=-1)
        rf_model_imp.fit(X_num_sample_rf_imputed, y_test_sample)
        num_importances = pd.Series(rf_model_imp.feature_importances_, index=num_features_for_rf)
        print("num_importances recalculada.")
    else:
        print("No hay variables numéricas para calcular num_importances.")
        num_importances = pd.Series()


# --- 8. Bloque de Comparación y Gráficos (usando la versión corregida de ayer) ---
if relev_ghost_output and not num_importances.empty:
    ghost_importance_raw = relev_ghost_output["relev_ghost"]
    ghost_importance_series = pd.Series(ghost_importance_raw, index=features_to_evaluate)
    print("\nRelevancia Bruta (Ghost Variables) - Solo > 0:")
    print(ghost_importance_series[ghost_importance_series > 0].sort_values(ascending=False))

    print("\nUsando num_importances para la comparación.")

    # 1. Filtrar las variables numéricas DE LA LISTA features_to_evaluate
    numeric_features_in_evaluation = [f for f in features_to_evaluate if f in var_numericas]

    # 2. Filtrar la serie ghost para esas variables numéricas
    ghost_importance_num_only = ghost_importance_series.reindex(numeric_features_in_evaluation).fillna(0)
    ghost_importance_num_only = ghost_importance_num_only[ghost_importance_num_only >= 0]

    # 3. Filtrar num_importances (la importancia RF del EDA) para esas mismas variables numéricas
    try:
        other_importance_raw = num_importances.reindex(ghost_importance_num_only.index).fillna(0)
    except AttributeError:
         print("Error: 'num_importances' no es una Serie de Pandas válida o falta .reindex.")
         other_importance_raw = pd.Series()
    except KeyError as e:
        print(f"Error de clave al reindexar 'num_importances': {e}")
        other_importance_raw = pd.Series()

    if not other_importance_raw.empty:
        # Normalizaciones...
        ghost_importance_norm = ghost_importance_num_only.copy()
        ghost_sum = ghost_importance_norm.sum()
        if ghost_sum > 0: ghost_importance_norm /= ghost_sum
        else: ghost_importance_norm = pd.Series(0.0, index=ghost_importance_num_only.index)

        other_importance_norm = other_importance_raw.copy()
        other_sum = other_importance_norm.sum()
        if other_sum > 0: other_importance_norm /= other_sum
        else: other_importance_norm = pd.Series(0.0, index=other_importance_raw.index)

        # DataFrame de comparación...
        comparison_df = pd.DataFrame({
            "RF Importance (EDA)": other_importance_norm,
            "Ghost Importance": ghost_importance_norm
        }).fillna(0)
        comparison_df = comparison_df.sort_values(by="Ghost Importance", ascending=False)
        print("\nComparación de Variable Importance (RF EDA vs Ghost Variables - Solo Numéricas)")
        print(comparison_df)

        # Gráfico...
        if not comparison_df.empty:
            plt.figure(figsize=(12, max(6, len(comparison_df)*0.3)))
            bar_width = 0.35
            x_pos = np.arange(len(comparison_df))
            plt.barh(x_pos + bar_width/2, comparison_df['RF Importance (EDA)'], bar_width, label='RF Importance (EDA)')
            plt.barh(x_pos - bar_width/2, comparison_df['Ghost Importance'], bar_width, label='Ghost Variables')
            plt.yticks(x_pos, comparison_df.index)
            plt.xlabel('Importancia Normalizada')
            plt.ylabel('Variables Numéricas')
            plt.title('Importancia de Variables Numéricas: RF (EDA) vs Variables Fantasma')
            plt.gca().invert_yaxis()
            plt.legend()
            plt.tight_layout()
            plt.show()
        else: print("No hay datos para graficar la comparación.")
    else: print("No se pudo alinear 'num_importances' con los resultados de ghost.")
elif num_importances.empty :
    print("num_importances está vacía, no se puede realizar la comparación.")
else:
    print("Hubo un error en el cálculo de la relevancia fantasma. No se puede continuar.")


# --- 9. Visualizaciones Adicionales (Líneas de Tendencia, Histograma de Relevancia) ---
# (Pega aquí el código para estas visualizaciones que te proporcioné antes,
#  usando A_matrix, relevance_series, evaluated_features del relev_ghost_output)

if relev_ghost_output:
    A_matrix_viz = relev_ghost_output['A']
    relevance_scores_viz = relev_ghost_output['relev_ghost']
    evaluated_features_viz = features_to_evaluate # La lista original que pasaste a relev_ghost_var
    relevance_series_viz = pd.Series(relevance_scores_viz, index=evaluated_features_viz)

    # Gráfica de Línea de Tendencia del Vector de Efectos (Matriz A)
    print("\n--- Gráficas de Línea de Tendencia del Vector de Efectos (Matriz A) ---")
    top_n_to_plot = 10
    # Seleccionar solo las columnas correspondientes a evaluated_features_viz en A_matrix_viz
    # y luego de esas, las top N por relevancia

    # Asegurarse de que solo se consideran las features para las que se calculó relevancia > 0
    valid_numeric_features_for_plot = relevance_series_viz[relevance_series_viz > 0].index.tolist()
    valid_numeric_features_for_plot = [f for f in valid_numeric_features_for_plot if f in var_numericas]


    if valid_numeric_features_for_plot:
        sorted_relevance_df = relevance_series_viz[valid_numeric_features_for_plot].sort_values(ascending=False)
        features_to_plot_names_viz = sorted_relevance_df.index[:top_n_to_plot].tolist()

        # Obtener los índices correctos de A_matrix_viz
        indices_in_A = [evaluated_features_viz.index(name) for name in features_to_plot_names_viz]
        A_matrix_to_plot_viz = A_matrix_viz[:, indices_in_A]

        num_samples_plot_viz = A_matrix_to_plot_viz.shape[0]
        num_features_plot_viz = A_matrix_to_plot_viz.shape[1]

        plt.figure(figsize=(15, 8))
        colors = plt.cm.get_cmap('tab10', num_features_plot_viz)

        for j_viz in range(num_features_plot_viz):
            plt.plot(np.arange(num_samples_plot_viz), A_matrix_to_plot_viz[:, j_viz], label=f'{features_to_plot_names_viz[j_viz]} (Relev: {relevance_series_viz[features_to_plot_names_viz[j_viz]]:.4f})', color=colors(j_viz), alpha=0.7)

        plt.title('Vector de Efectos (y_hat_base - y_hat_ghost) por Muestra (Top Numéricas)')
        plt.xlabel('Índice de Muestra (en X_test_sample)')
        plt.ylabel('Cambio en la Predicción de Fuga (Probabilidad)')
        plt.legend(loc='best', bbox_to_anchor=(1.05, 1), borderaxespad=0.)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.tight_layout(rect=[0, 0, 0.85, 1])
        plt.show()

        # Histograma de la Variabilidad ("Puntuaciones de Relevancia Fantasma")
        print("\n--- Histograma de las Puntuaciones de Relevancia Fantasma ---")
        valid_relevance_scores_viz = relevance_series_viz[valid_numeric_features_for_plot] # Solo numéricas con score
        if not valid_relevance_scores_viz.empty:
            plt.figure(figsize=(10, 6))
            sns.histplot(valid_relevance_scores_viz, kde=True, bins=max(10, len(valid_relevance_scores_viz)//2))
            plt.title('Distribución de las Puntuaciones de Relevancia Fantasma (Variables Numéricas)')
            plt.xlabel('Puntuación de Relevancia (Varianza del Efecto)')
            plt.ylabel('Frecuencia')
            plt.grid(True, linestyle='--', alpha=0.7)
            plt.show()
            print("\nEstadísticas Descriptivas de las Puntuaciones de Relevancia:")
            print(valid_relevance_scores_viz.describe())
        else:
            print("No hay puntuaciones de relevancia válidas (numéricas > 0) para graficar.")

        # Distribución de los Efectos Individuales (Columnas de A)
        print("\n--- Distribución de los Efectos Individuales (Columnas de A - Top Numéricas) ---")
        df_A_to_plot_viz = pd.DataFrame(A_matrix_to_plot_viz, columns=features_to_plot_names_viz)
        if not df_A_to_plot_viz.empty:
            plt.figure(figsize=(15, max(8, num_features_plot_viz*1.5))) # Ajustar altura dinámicamente
            cols_subplot = 2
            rows_subplot = (num_features_plot_viz + cols_subplot - 1) // cols_subplot

            for i_viz, feature_name_viz in enumerate(features_to_plot_names_viz):
                plt.subplot(rows_subplot, cols_subplot, i_viz + 1)
                sns.histplot(df_A_to_plot_viz[feature_name_viz], kde=True, bins=30)
                plt.title(f'Distribución del Efecto para: {feature_name_viz}')
                plt.xlabel('Cambio en Predicción')
                plt.ylabel('Frecuencia')
            plt.tight_layout()
            plt.show()
        else:
             print("No hay datos para graficar la distribución de efectos.")
    else:
        print("No hay variables numéricas con relevancia > 0 para graficar.")

# --- 10. Guardar los Resultados del Análisis Fantasma ---
if 'relev_ghost_output' in locals() and relev_ghost_output is not None and 'features_to_evaluate' in locals():
    #save_folder_ghost = 'ghost_analysis_data' # Para guardar local
    save_folder_ghost = '/content/drive/MyDrive/CDD/Sem03/PGII/fuga-clientes-bc-v2-ghost-v/ghost_analysis_data' # para guardar en el drive.
    os.makedirs(save_folder_ghost, exist_ok=True)
    data_to_save_ghost = {
        'A_matrix': relev_ghost_output["A"],
        'relevance_scores': relev_ghost_output["relev_ghost"],
        'evaluated_features': features_to_evaluate,
        'base_predictions': relev_ghost_output["y_hat_base"]
    }
    file_path_ghost = os.path.join(save_folder_ghost, 'ghost_variable_analysis_results.pkl')
    try:
        with open(file_path_ghost, 'wb') as f: pickle.dump(data_to_save_ghost, f)
        print(f"Resultados del análisis fantasma guardados en: {file_path_ghost}")
    except Exception as e: print(f"Error al guardar resultados fantasma: {e}")
```

```python
# ============== VISUALIZACIONES ADICIONALES DE VARIABLES FANTASMA ==============

# Asegurarnos de que las variables base existen
if 'relev_ghost_output' not in locals() or relev_ghost_output is None:
    print("Error: 'relev_ghost_output' no está definido o es None. Ejecuta el cálculo de variables fantasma primero.")
    # Aquí podrías añadir un `raise Exception` o simplemente no continuar con los plots.
elif 'features_to_evaluate' not in locals():
    print("Error: 'features_to_evaluate' no está definido.")
elif 'var_numericas' not in locals(): # Tu lista de nombres de variables numéricas
    print("Error: 'var_numericas' no está definida.")
else:
    print("Procediendo con visualizaciones adicionales...")

    # Extraer datos necesarios de relev_ghost_output
    A_matrix = relev_ghost_output['A']
    V_matrix = relev_ghost_output['V']
    # 'relevance_scores' está en relev_ghost_output['relev_ghost']
    # 'evaluated_features' es tu 'features_to_evaluate'

    # Crear (o reconfirmar) la Serie de relevancia
    relevance_series = pd.Series(relev_ghost_output['relev_ghost'], index=features_to_evaluate)

    # --- Preparar listas de features numéricas que SÍ fueron procesadas y tienen score ---
    # (Esto es importante porque relev_ghost_var omite las categóricas)

    # 1. Nombres de las features numéricas que estaban en la lista de evaluación original
    numeric_features_that_were_in_evaluation = [f for f in features_to_evaluate if f in var_numericas]

    # 2. De esas, las que realmente obtuvieron un score de relevancia (deberían ser todas las numéricas)
    #    y las que están en la serie de relevancia (para consistencia)
    processed_numeric_features = [
        f for f in numeric_features_that_were_in_evaluation
        if f in relevance_series.index and relevance_series[f] >= 0 # O > 0 si quieres excluir las de efecto nulo
    ]

    # 3. Obtener los índices correspondientes en A_matrix y V_matrix
    #    Estos índices se basan en el orden de 'features_to_evaluate'
    numeric_indices_in_A_V = [features_to_evaluate.index(feat) for feat in processed_numeric_features]

    if not processed_numeric_features:
        print("No hay variables numéricas procesadas con relevancia para visualizar en detalle.")
    else:
        print(f"Visualizando detalles para {len(processed_numeric_features)} variables numéricas: {processed_numeric_features}")

        # --- 1. Heatmap de la Matriz V (para las numéricas procesadas) ---
        if len(processed_numeric_features) > 0:
            V_numeric_subset = V_matrix[np.ix_(numeric_indices_in_A_V, numeric_indices_in_A_V)]

            plt.figure(figsize=(max(10, len(processed_numeric_features)*0.7), max(8, len(processed_numeric_features)*0.6)))
            sns.heatmap(V_numeric_subset, annot=False, cmap="viridis",
                        xticklabels=processed_numeric_features, yticklabels=processed_numeric_features)
            plt.title('Heatmap de la Matriz de Relevancia V (Variables Fantasma - Numéricas Procesadas)')
            plt.xticks(rotation=90)
            plt.yticks(rotation=0)
            plt.tight_layout()
            plt.show()
        else:
            print("No hay suficientes variables numéricas procesadas para el heatmap de V.")

        # --- 2. PCA de la Matriz A (para las numéricas procesadas) ---
        if len(processed_numeric_features) > 1: # PCA necesita al menos 2 features
            A_numeric_subset = A_matrix[:, numeric_indices_in_A_V]

            from sklearn.decomposition import PCA # Asegurar importación
            # Opcional: escalar A_numeric_subset si es necesario
            # from sklearn.preprocessing import StandardScaler
            # scaler_A = StandardScaler()
            # A_scaled_subset = scaler_A.fit_transform(A_numeric_subset)
            # pca_A = PCA(n_components=min(5, A_numeric_subset.shape[1])).fit(A_scaled_subset)

            pca_A = PCA(n_components=min(5, A_numeric_subset.shape[1])).fit(A_numeric_subset)

            # Scree plot
            plt.figure(figsize=(10, 6))
            plt.bar(range(1, pca_A.n_components_ + 1), pca_A.explained_variance_ratio_, color='skyblue')
            plt.xlabel('Componente Principal del Vector de Efectos A')
            plt.ylabel('Ratio de Varianza Explicada')
            plt.title('Scree Plot de PCA sobre la Matriz de Efectos A (Numéricas Procesadas)')
            plt.xticks(range(1, pca_A.n_components_ + 1))
            plt.grid(axis='y', linestyle='--', alpha=0.7)
            plt.show()

            # Cargas de los componentes
            loadings_df = pd.DataFrame(pca_A.components_.T,
                                       columns=[f'PC{i+1}' for i in range(pca_A.n_components_)],
                                       index=processed_numeric_features)
            print("\nCargas de los Componentes Principales de A (Numéricas Procesadas):")
            print(loadings_df.round(3))

            # Graficar las cargas de los primeros 2 PCs
            if pca_A.n_components_ >= 2:
                plt.figure(figsize=(10, 8))
                plt.scatter(loadings_df['PC1'], loadings_df['PC2'], color='dodgerblue', alpha=0.7)
                for i, txt in enumerate(loadings_df.index):
                    plt.annotate(txt, (loadings_df['PC1'][i], loadings_df['PC2'][i]),
                                 textcoords="offset points", xytext=(0,5), ha='center', fontsize=9)
                plt.xlabel('Carga PC1 (Matriz A)')
                plt.ylabel('Carga PC2 (Matriz A)')
                plt.title('Cargas de Variables en los Primeros 2 PCs de A (Numéricas Procesadas)')
                plt.axhline(0, color='grey', lw=1, linestyle='--')
                plt.axvline(0, color='grey', lw=1, linestyle='--')
                plt.grid(True, linestyle=':', alpha=0.7)
                plt.show()
        elif len(processed_numeric_features) == 1:
             print("Solo hay una variable numérica procesada, no se puede aplicar PCA.")
        else:
             print("No hay suficientes variables numéricas procesadas para aplicar PCA a la matriz A.")


        # --- 3. Boxplots Agrupados de las Columnas de A (para las numéricas procesadas) ---
        if len(processed_numeric_features) > 0:
            A_numeric_df = pd.DataFrame(A_matrix[:, numeric_indices_in_A_V], columns=processed_numeric_features)

            # Ordenar por relevancia para mejor visualización
            sorted_features_for_boxplot = relevance_series[processed_numeric_features].sort_values(ascending=False).index
            A_numeric_df_sorted = A_numeric_df[sorted_features_for_boxplot]

            plt.figure(figsize=(max(12, len(processed_numeric_features)*0.6) , 7)) # Ajustado
            sns.boxplot(data=A_numeric_df_sorted, orient="v", palette="viridis")
            plt.title('Distribución de los Vectores de Efecto (A_j) por Variable (Numéricas Procesadas)')
            plt.ylabel('Cambio en la Predicción de Fuga')
            plt.xlabel('Variables')
            plt.xticks(rotation=90)
            plt.grid(axis='y', linestyle='--', alpha=0.7)
            plt.tight_layout()
            plt.show()
        else:
            print("No hay variables numéricas procesadas para mostrar boxplots de efectos.")

        # --- 4. Gráfico de Relevancia Fantasma vs. Importancia RF (Scatter Plot) ---
        #    (Asumiendo que 'comparison_df' se calculó correctamente antes y está disponible)
        if 'comparison_df' in locals() and not comparison_df.empty:
            # Asegurarse que comparison_df solo tiene las processed_numeric_features
            comparison_df_numeric = comparison_df.reindex(processed_numeric_features).dropna()

            if not comparison_df_numeric.empty:
                plt.figure(figsize=(10, 8))
                sns.scatterplot(data=comparison_df_numeric, x='RF Importance (EDA)', y='Ghost Importance', color='green', alpha=0.7)
                for i in range(comparison_df_numeric.shape[0]):
                    plt.text(x=comparison_df_numeric['RF Importance (EDA)'].iloc[i]+0.001,
                             y=comparison_df_numeric['Ghost Importance'].iloc[i]+0.001,
                             s=comparison_df_numeric.index[i],
                             fontdict=dict(color='black',size=9))
                plt.title('Relevancia Fantasma vs. Importancia RF (Numéricas Procesadas)')
                plt.xlabel('Importancia Normalizada RF (EDA)')
                plt.ylabel('Importancia Normalizada Ghost Variables')
                plt.grid(True, linestyle=':', alpha=0.7)
                plt.show()

                correlation = comparison_df_numeric['RF Importance (EDA)'].corr(comparison_df_numeric['Ghost Importance'])
                print(f"Correlación entre Importancia RF y Ghost (Numéricas Procesadas): {correlation:.2f}")
            else:
                print("No hay datos en comparison_df para las variables numéricas procesadas.")
        else:
            print("'comparison_df' no está definido o está vacío. Omite el scatter plot de comparación.")


        # --- 5. Gráfico de "Impacto Acumulado" o "Curva de Relevancia" ---
        if len(processed_numeric_features) > 0:
            # Usar relevance_series filtrada solo para las numéricas procesadas
            valid_relevance_scores_viz = relevance_series[processed_numeric_features].sort_values(ascending=False)

            if not valid_relevance_scores_viz.empty and valid_relevance_scores_viz.sum() > 0:
                cumulative_relevance = np.cumsum(valid_relevance_scores_viz) / valid_relevance_scores_viz.sum()

                plt.figure(figsize=(12, 7)) # Ajustado
                plt.plot(range(1, len(cumulative_relevance) + 1), cumulative_relevance, marker='o', linestyle='-', color='purple')
                plt.title('Relevancia Fantasma Acumulada (Variables Numéricas Procesadas)')
                plt.xlabel('Número de Variables (Ordenadas por Relevancia)')
                plt.ylabel('Porcentaje de Relevancia Total Acumulada')
                plt.xticks(ticks=range(1, len(cumulative_relevance) + 1), labels=valid_relevance_scores_viz.index, rotation=90, ha='right')
                plt.yticks(np.arange(0, 1.1, 0.1))
                plt.grid(True, linestyle='--', alpha=0.7)
                plt.axhline(0.8, color='red', linestyle='--', label='80% Relevancia')
                plt.axhline(0.9, color='orange', linestyle='--', label='90% Relevancia')
                plt.legend()
                plt.tight_layout()
                plt.show()

                n_for_80 = (cumulative_relevance >= 0.80).argmax() + 1
                print(f"{n_for_80} variables explican al menos el 80% de la relevancia fantasma total (entre las numéricas procesadas).")
            elif not valid_relevance_scores_viz.empty and valid_relevance_scores_viz.sum() == 0:
                 print("Todas las relevancias fantasmas numéricas son cero. No se puede graficar la curva acumulada.")
            else:
                print("No hay puntuaciones de relevancia válidas para la curva acumulada.")
```
